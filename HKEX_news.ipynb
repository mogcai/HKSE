```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
start_date='20240513'
end_date='20240913'
stock_id='12593'
url='https://www1.hkexnews.hk/search/titlesearch.xhtml'
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0',
         }

data=f'lang=ZH&market=SEHK&searchType=0&documentType=&t1code=&t2Gcode=&t2code=&stockId={stock_id}&from={start_date}&to={end_date}&category=0&title='
headers={'Content-Type': 'application/x-www-form-urlencoded',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0'}
r=requests.post(url, data=data, headers=headers)
r
```

```python
def get_dict_record(record):
    stock_no=record.find(lambda tag: tag.name == "td" 
                    and tag.find(string=re.compile(r'股份代號')))\
                        .get_text(strip=True).split(':')[-1]

    stock_name=record.find(lambda tag: tag.name == "td" 
                    and tag.find(string=re.compile(r'股份簡稱')))\
                        .get_text(strip=True).split(':')[-1]
    title=record.find('div', {'class': 'headline'}).get_text().strip()
    domain='https://www1.hkexnews.hk'
    link=domain+record.find('div', {'class': 'doc-link'}).a.get('href')
    doc_name=record.find('div', {'class': 'doc-link'}).a.get_text().strip()
    pub_date=record.find(lambda tag: tag.name == "td" 
                    and tag.find(string=re.compile(r'發放時間')))\
                        .get_text(strip=True).split('間:')[-1]
    dict_record={
        'stock_no': stock_no,
        'stock_name': stock_name,
        'title': title,
        'doc_name': doc_name,
        'link': link, 
        'pub_date': pub_date
    }
    return dict_record
```

```python
soup=BeautifulSoup(r.content)
records=soup.find('main', {'class': 'title-search'}).find_all('tr')
lst_record=[get_dict_record(record) for record in records[1:]]
lst_record
```
